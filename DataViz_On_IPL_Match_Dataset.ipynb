# <center>  Data Vizualisation on the IPL Match Dataset 

### Importing important libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy as sc
```

### Importing the dataset

```python
df = pd.read_csv('innings_deliveries.csv')
df.head()
```

### Dataset exploration

```python
df.info()
```

```python
#For float variables
df.describe()
```

```python
#For string variables
df.describe(include='object')
```

### Data Vizualisation

```python
plt.figure(figsize=(10,6))
sns.histplot(x='over',kde=True, cumulative=False ,data=df)
plt.xlabel('Over variable')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(y='runs_total',kde=True, cumulative=False ,data=df)
plt.xlabel('Total runs')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='team', orient="h", data=df)#stat=['proportion'],
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='wicket_kind', order=['bowled','lbw','run out', 'caught'], data=df)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='wicket_kind', hue='team', order=['bowled','lbw','run out', 'caught'], data=df)
```

```python
df.columns
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y= 'runs_batter',data=df)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(y= 'runs_total',hue= 'bowler', data=df)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='runs_batter',data=df)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(y='runs_total', hue='team',data=df)
plt.show()
```
